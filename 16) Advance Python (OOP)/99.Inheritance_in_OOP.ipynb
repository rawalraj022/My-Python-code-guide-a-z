# Inheritance in OOP :

Inheritance allows us to define a class that inherits all the methods and properties from another class. Parent class is the class being inherited from, also called base class. Child class is the class that inherits from another class, also called derived class.

But there is a condition that it should be noe private attribute and non private methods. Which means we can only inherit the non private attribute and non private methods. Which is we have also learned in aggregation concept. 

If the parent class is having private attribute and private methods then we cannot access the parent class property in child class. So, we have to make sure that parent class should not have private attribute and private methods. 

But if we want to access the private attribute and private methods in child class then we have to use getter and setter method.

- non private attribute
- non private methods

```python
## Q) Let's say we are implementing coursera website and we have two types of users, Student and Instructor.
    ## Both Student and Instructor have common methods like login and register.
    ## Student has additional method enroll and Instructor has additional method create.
    ## Implement this using inheritance.

class User: #parent class
    #constructor
    def __init__(self): 
        self.name = "Rajkumar"
        self.gender = "Male"

    # login and register are common methods
    def login(self):
        print("Login Done!")

    def register(self):
        print("register Done!")


## If I want to create a child class for Student , I have to inherit the parent class User 
class Student(User): #child class  # Inheritance
    def __init__(self): #constructor
        self.rollno = 12

    def enroll(self):  #additional method for student class only
        print("Enrolled the course!")


## If I want to create a child class for Instructor , I have to inherit the parent class User
class Instructor(User): #child class  # Inheritance
    def __init__(self):
        self.idno = 120

    def create(self):
        print("created the course!")

```

```python
## defining object for Student class
s = Student() #object creation for Student class
```

```python

s.enroll() #calling the enroll method of Student class
s.login() #calling the login method of User class
s.register() #calling the register method of User class
```

```python

i = Instructor() #object creation for Instructor class 
i.create() #calling the create method of Instructor class
i.login() #calling the login method of User class
i.register() #calling the register method of User class
```


It will throw one error when it will tell 'Student' object has no attribute 'name'. Why because I told you I can access my parent class property , non-private property but why it is not able to access the name variable or name attribute. Why because there is a concept inside inheritance . In both classes we are having the constructor , so whenever we initialize my student object that means student class definitely constructor inside student class will get executed by default. It will always follow one rule , the rule is if one class is having his own constructor then that constructor would be executed very first. So, That means it will give the first priority to it's constructor. So, whenever I'm creating the student object , it will give the first priority to student class constructor and it will not give the first priority to user class constructor . Then, it will give the priority to the second constructor which is it is getting inherited. 

**Now what is happening here, whenever I'm creating the student object , it will give the first priority to student class constructor and it will not give the first priority to user class constructor. So, that means the name variable is not getting initialized. So, that's why it is throwing the error.** 

So, to resolve this error we have to use the **super() method.** **super() method is used to call the parent class constructor.** So, whenever we are using the super() method , it will call the parent class constructor and it will initialize the parent class property.
