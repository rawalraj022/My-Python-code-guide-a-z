#### Class Vs Object :

**Class** : Class is a blueprint for the object. It is a template for the object. It defines the properties and behaviors of the object. It is a logical entity. It is a logical entity that has some attributes and methods.

**Object** : Object is an instance of the class. Whenever we create an object, it will occupy some memory. It is a physical entity. It is a physical entity that has some attributes and methods.


#### Syntax :


#### Example :

Let's see an example to understand the class and object in OOP.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name : ", self.name)
        print("Age : ", self.age)

# Creating an object of the Student class
s1 = Student("John", 25)
s1.display()
```

In the above example, we have created a class called Student. It has two attributes called name and age. It has two methods called __init__ and display. We have created an object of the Student class called s1. We have passed the values of the attributes name and age while creating the object. We have called the display method using the object s1.

- attributes means the properties of the object. methods means the behaviors of the object.

- __init__ is a constructor. It is a special method in python. It is used to initialize the attributes of the object. It is called when we create an object of the class. It is called automatically when we create an object of the class. It is a magic method in python. 

- display is a method. It is used to display the attributes of the object. It is called using the object of the class. It is a instance method. It is a public method. It is a method that is accessible outside the class.

- s1 is an object of the Student class. It is an instance of the Student class. It is a physical entity that has some attributes and methods. It occupies some memory. It is a physical entity that has some properties and behaviors.

- John is the value of the attribute name. 25 is the value of the attribute age.

Whenever we are creating our own custom class, that class have two properties :

# Class Basic Structure :

Whenever we'll be defining a class inside python, we have to use the keyword called **class** followed by the class name. The structure of the class is as follows :

The syntax of the class is as follows :
```python
class ClassName:
    # Data or property
    # Functions or behavior
```

Example : 
```python
class Car:
    # Data or property
    color = "Red"
    model = "2021"
    company = "BMW"
    seats = 4
    wheels = 4

    # Functions or behavior
    def start(self):
        print("Car is started")

    def stop(self):
        print("Car is stopped")

    def drive(self):
        print("Car is driven")
```






# Object is an Instance of the Class :
