Whenever we are defining inheritance class diagram , that time **we have to use arrow symbol.** **Instead of diamond symbol.** Whenever we see this arrow symbol that means its a inheritance class diagram and concept will remain same. It will have our data property and method property and if we see the **+** symbol that mean its a **public property** and if we see the **-** symbol that means its a **private property**. That how we create our inheritance class diagram. 

**Example of Inheritance :**

```python
## Using _Address__city to access the private attribute __city from the outside of the class.
    # We can access the private attribute __city from the outside of the class by using _Address__city.
    # We have used _Address__city to access the private attribute __city from the outside of the class instead of get_city method.

class Customer:
    def __init__(self,name,gender,address):
        self.name = name
        self.gender = gender
        self.address = address

    def print_info(self):
        print(f"Name: {self.name}, Gender: {self.gender}, Address: {self.address._Address__city}, {self.address.pin}, {self.address.state}")
        ## Using _Address__city to access the private attribute __city from the outside of the class instead of get_city method.

class Address:
    def __init__(self, city, pin, state):
        self.__city = city
        self.pin = pin
        self.state = state

```



In the inheritance class diagram, we have two blocks. One is the parent class block and another is the child class block.

In the parent class block, There are three small blocks ,

1. First block contain the name of the class which is customer class.
2. Second block contain the name of the attributes or variables which are name, gender, address. these all are public properties because I'm not using any private properties here. So, I'm going to give the plus sign in front of the attributes.
3. Third block contain the name of the methods which are print_info method. This is also a public method because I'm not using any private method here. So, I'm going to give the plus sign in front of the method.

In the child class block, There are three small blocks ,

1. First block contain the name of the class which is address class.
2. Second block contain the name of the attributes or variables which are city, pin, state. Here, city is a private property because I'm using double underscore in front of the city. So, I'm going to give the minus sign in front of the city and pin, state are public properties because I'm not using any private properties here. So, I'm going to give the plus sign in front of the pin and state.
3. Third block contain the name of the methods which are **init** method. This is also a public method because I'm not using any private method here. So, I'm going to give the plus sign in front of the method.

If you want to show inheritance in class diagram, you can use a **arrow shape** at the parent class and a line with an arrow at the child class.

**Whoever having this particular Arrow shape, that means this particular class owns the other class.** This is called inheritance diagram in OOP.

We have to make a habit to draw these kinds of diagrams whenever we are trying to create the class or we are performing inheritance or composition or anything else. Just try to draw the diagram as well because this will help us a lot to understand our software or system or architecture. How we have designed our software or system or architecture.
