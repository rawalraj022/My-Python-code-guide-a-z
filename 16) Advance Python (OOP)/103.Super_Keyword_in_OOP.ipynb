
# Super Keyword in OOP Python :

Constructor overwriting is a concept in OOP where a subclass can have its own constructor and can also call the constructor of the parent class. **This is done using the super() function.** The super() function is used to call the constructor of the parent class. 
Otherwise, first of all it will call the constructor of the child class and then the constructor of the parent class.

Method overwriting is a concept in OOP where a subclass can have its own method and can also call the method of the parent class. **This is done using the super() function.** The super() function is used to call the method of the parent class.
Otherwise, first of all it will call the method of the child class and then the method of the parent class.

**By using the super() function, we can call the constructor and method of the parent class directly without it going through the child class first and then the parent class.**




```python
##Q) How to call a method of the parent class from the child class in Python?
##Ans) We can call a method of the parent class from the child class using the super() function.
## The super() function is used to call the method of the parent class from the child class.
## The super() function returns a temporary object of the superclass that allows us to call the method of the superclass.
## The super() function can be used to call the constructor of the parent class from the child class.


class Phone: ## Parent class
    ## Constructor of the class Phone
    def __init__(self, price, brand, camera):
        print("Inside Phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera
    
    ## get_price method to get the price of the phone
    def get_price(self):
        return self.__price
    
    ## buy method to print a message
    def buy(self):
        print("Buying a Phone inside Phone constructor")

## Class SmartPhone inherits the class Phone
class SmartPhone(Phone): ## Child class
    def buy(self):
        print("Buying a Phone inside SmartPhone constructor")
        
        super().buy() ## Calling the buy method of the parent class Phone inside the child class SmartPhone 
    

## Creating an object of the class SmartPhone with the price, brand and camera as the arguments
S = SmartPhone(20000, "Apple", 12) ## This will call the constructor of the class Phone as well as the class SmartPhone as the class SmartPhone inherits the class Phone 

## Calling the buy method of the class SmartPhone which will call the buy method of the class Phone
S.buy()


## Explanation of the above code ? 

## We have created two classes Phone and SmartPhone. The class Phone is the parent class and the class SmartPhone is the child class.
## The class Phone has a constructor __init__ which takes three arguments price, brand, and camera.
## The class Phone has two methods get_price and buy.
## The class SmartPhone inherits the class Phone.
## The class SmartPhone has a method buy which prints a message Buying a Phone inside SmartPhone constructor.
## Inside the buy method of the class SmartPhone, we have called the buy method of the class Phone using
## super().buy(). The super() function returns a temporary object of the superclass that allows us to call the method of the superclass.
## The super().buy() will call the buy method of the class Phone.
## We have created an object S of the class SmartPhone with the price, brand, and camera as the arguments.
## When we call the buy method of the class SmartPhone using S.buy(), it will call the buy method of the class SmartPhone
## and then it will call the buy method of the class Phone using
## super().buy() inside the buy method of the class SmartPhone.
## The output of the above code will be:
    ## Buying a Phone inside SmartPhone constructor
    ## Buying a Phone inside Phone constructor

## As we can see, the buy method of the class SmartPhone is called first and then the buy method of the class Phone is called.

```

```python
##Q) How to call the attribute of the parent class from the child class in Python?
##Ans) We can call the attribute of the parent class from the child class using the super() function.
## The super() function is used to call the attribute of the parent class from the child class.
## The super() function returns a temporary object of the superclass that allows us to call the attribute of the superclass.
## The super() function can be used to call the attribute of the parent class from the child class.


## Class Phone with the constructor __init__ which takes three arguments price, brand, and camera
class Phone: 
    ## Constructor of the class Phone
    def __init__(self, price, brand, camera):  ## price, brand and camera are the attributes of the class Phone
        print("Inside Phone constructor")      ## This will print Inside Phone constructor
        self.price = price                     ## Assigning the value of price to the attribute price of the class Phone
        self.brand = brand                     ## Assigning the value of brand to the attribute brand of the class Phone
        self.camera = camera                   ## Assigning the value of camera to the attribute camera of the class Phone


## Class SmartPhone inherits the class Phone 
class SmartPhone(Phone):
    ## Constructor of the class SmartPhone
    def __init__(self, price, brand, camera, os, ram): ## price, brand, camera, os and ram are the attributes of the class SmartPhone
        print("Inside SmartPhone constructor")         ## This will print Inside SmartPhone constructor
        super().__init__(price, brand, camera)         ## Calling the constructor of the parent class Phone using super() function 
        self.os = os                                   ## Assigning the value of os to the attribute os of the class SmartPhone
        self.ram = ram                                 ## Assigning the value of ram to the attribute ram of the class SmartPhone


S = SmartPhone(20000, "Samsung", 3, "Android", 16)    ## Creating an object of the class SmartPhone with the price, brand, camera, os and ram as the arguments 
print(S.brand)                                    ## Accessing the attribute brand of the class Phone using the object S of the class SmartPhone
print(S.os)                                       ## Accessing the attribute os of the class SmartPhone using the object S of the class SmartPhone
print(S.price)                                   ## Accessing the attribute price of the class Phone using the object S of the class SmartPhone 


## Explanation of the above code ?

## We have created two classes Phone and SmartPhone. The class Phone is the parent class and the class SmartPhone is the child class.
## The class Phone has a constructor __init__ which takes three arguments price, brand, and camera.
## The class SmartPhone inherits the class Phone.
## The class SmartPhone has a constructor __init__ which takes five arguments price, brand, camera, os, and ram.
## Inside the constructor of the class SmartPhone, we have called the constructor of the class Phone using
## super().__init__(price, brand, camera). The super() function returns a temporary object of the superclass that allows us to call the constructor of the superclass.
## The super().__init__(price, brand, camera) will call the constructor of the class Phone.
## We have created an object S of the class SmartPhone with the price, brand, camera, os, and ram as the arguments.
## We have accessed the attribute brand of the class Phone using
## S.brand, the attribute os of the class SmartPhone using S.os, and the attribute price of the class Phone using S.price.
## The output of the above code will be:
    ## Inside SmartPhone constructor
    ## Inside Phone constructor
    ## Samsung
    ## Android
    ## 20000

## As we can see, the constructor of the class SmartPhone is called first, then the constructor of the class Phone is called.
## We have accessed the attribute brand of the class Phone using
## S.brand, the attribute os of the class SmartPhone using S.os, and the attribute price of the class Phone using S.price.

```
