````python

class AtmMachine:            # AtmMachine is a class name and where ATM and Machine are two words and M is capital letter of Machine and A is capital letter of ATM . 
    ## First of all we have to define data properties of Atm Machine inside the AtmMachine class.
        ## Data properties are the properties which are used to store data in the class.
        ## Data properties are also called as attributes of the class.
    ## Data properties are defined inside the class but outside the methods. 
    
    ## Q) How to define data properties in the class? 
        ## => To define data properties in the class,for this we need to use one special 
            # function inside a class which is called as Constructor. 
                ## Q) What is Constructor? 
                    ## => Constructor is a special function which is used to initialize the data properties of the class.
                    ## Whenever we see __init__ function inside a class, it means it is a constructor function.
                        # Special function has some super powers.
    ## Whatever data property we want to write , 
        # we have to write inside the constructor function.
    ## We need three data properties like pin number, balance and withdraw. 
    
    ## Constructor (special function) -> super power -> initialize the data properties of the class.
    
    def __init__(self):     # Constructor function is defined inside the class.
        self.pin = ""      # self is a keyword which is used to define the data properties of the class and it is used to access the data properties of the class.
        self.balance = 0   
        self.withdraw = 0  
        self.menu()        # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
        
    ## Now we have to define the methods of the class.
        ## Methods are the functions which are defined inside the class.
        ## Methods are used to perform some operations on the data properties of the class.
    
    ## We need to define 4 methods for the AtmMachine class. 
        ## 1) check_balance
        ## 2) withdraw 
        ## 3) pin number
        
    ## Now we have to define the menu method.
    def menu(self):                    # menu is a method of the AtmMachine class and self is a keyword which is used to access the data properties of the class.
        user_input = input(          # input is a function which is used to take input from the user and it is used to take input in the form of string.
            """
            Hi how can I help you ?
            1. Press 1 to create pin 
            2. Press 2 to change pin
            3. Press 3 to check balance
            4. Press 4 to withdraw money
            5. Anything to exit 
        
            """
        )
        
        ## Now we have to write the logic for the above options.
        if user_input == "1":          # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            ## create pin
            self.create_pin()          # self is a keyword which is used to access the data properties of the class and create_pin is a method of the AtmMachine class.
        
        elif user_input == "2":       # elif is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the elif
            ## change pin
            self.change_pin()          # self is a keyword which is used to access the data properties of the class and change_pin is a method of the AtmMachine class.
        
        elif user_input == "3":      # elif is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the elif
            ## check balance
            self.check_balance()      # self is a keyword which is used to access the data properties of the class and check_balance is a method of the AtmMachine class.
        
        elif user_input == "4":       # elif is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the elif
            ## withdraw money
            self.withdraw_balance()    # self is a keyword which is used to access the data properties of the class and withdraw_balance is a method of the AtmMachine class.
        
        else:                         # else is a keyword which is used to execute the block of code if the condition is false.
            exit()
            
    
    ## Now we have to define the create_pin method.
    def create_pin(self):              # create_pin is a method of the AtmMachine class and self is a keyword which is used to access the data properties of the class.
        ## Now we have to take the pin from the user.
        user_pin = input("Enter your pin: ")     # input is a function which is used to take input from the user and it is used to take input in the form of string.
        self.pin = user_pin
        
        ## Now we have to take the balance from the user.
        user_balance = int(input("Enter Balance:"))   # input is a function which is used to take input from the user and it is used to take input in the form of string. So we have to convert it into integer.
        self.balance = user_balance                 # self is a keyword which is used to define the data properties of the class and it is used to access the data properties of the class.
        
        print("Pin Created Successfully")          # print is a function which is used to print the message on the screen.
        self.menu()                               # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
    
    ## Now we have to define the change_pin method.
    def change_pin(self):
        old_pin = input("Enter your old pin: ")    # input is a function which is used to take input from the user and it is used to take input in the form of string.
        
        if old_pin == self.pin:                    # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            new_pin = input("Enter your new pin: ")   # input is a function which is used to take input from the user and it is used to take input in the form of string.
            self.pin = new_pin                        # self is a keyword which is used to access the data properties of the class.
            print("Pin Changed Successfully")         # print is a function which is used to print the message on the screen and it is used to print the message on the screen.
            self.menu()                              # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
        else:                                         # else is a keyword which is used to execute the block of code if the condition is false.
            print("Invalid Pin")                      # print is a function which is used to print the message on the screen.
            self.menu()                               # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
    
    ## Now we have to define the check_balance method.
    
    def check_balance(self):
        user_pin = input("Enter your pin: ")         # input is a function which is used to take input from the user and it is used to take input in the form of string.
        if user_pin == self.pin:                      # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            print(f"Your balance is {self.balance}")  # print is a function which is used to print the message on the screen and it is used to print the message on the screen.
        
        else:                                         # else is a keyword which is used to execute the block of code if the condition is false.
            print("Your pin is incorrect, please try again")            # print is a function which is used to print the message on the screen.
        self.menu()
    
    ## Now we have to define the withdraw method.
    
    def withdraw_balance(self):                       # withdraw_balance is a method of the AtmMachine class and self is a keyword which is used to access the data properties of the class.
        user_pin = input("Enter your pin: ")          # input is a function which is used to take input from the user and it is used to take input in the form of string.
        if user_pin == self.pin:                      # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            # allow to withdraw
            amount = int(input("Enter the amount : "))   # input is a function which is used to take input from the user and it is used to take input in the form of string. So we have to convert it into integer.
            if amount <= self.balance:                                       # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
                self.balance = self.balance - amount                         # self is a keyword which is used to access the data properties of the class and it is used to define the data properties of the class.
                print(f"You have withdrawn {amount}. Your new balance is {self.balance}")          # print is a function which is used to print the message on the screen.
            
            else:                                                            # else is a keyword which is used to execute the block of code if the condition is false.
                print("Insufficient balance!")                                # print is a function which is used to print the message on the screen.
        
        else:                                                                # else is a keyword which is used to execute the block of code if the condition is false and it is used to execute the block of code if the condition is false.
            print("Your pin is incorrect, please try again")                 # print is a function which is used to print the message on the screen.
        self.menu()                                                         # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
        
        

## Explanation of the above code:

    ## First of all we have to define the AtmMachine class.
    ## Inside the AtmMachine class, we have to define the data properties of the class.
        ## Data properties are the properties which are used to store data in the class.
        ## Data properties are also called as attributes of the class.
        ## Data properties are defined inside the class but outside the methods.
        ## We need three data properties like pin number, balance and withdraw.
        ## To define data properties in the class, we have to use one special function inside a class which is called as Constructor.
            ## Constructor is a special function which is used to initialize the data properties of the class.
            ## Whenever we see __init__ function inside a class, it means it is a constructor function.
    ## After defining the data properties of the class, we have to define the methods of the class which are used to perform some operations on the data properties of the class.
    ## We need to define 4 methods for the AtmMachine class.
        ## 1) check_balance
        ## 2) withdraw
        ## 3) pin number
        ## 4) menu
    ## Now we have to define the menu method.
        ## menu is a method of the AtmMachine class.
        ## Inside the menu method, we have to take the input from the user and we have to write the logic for the above options.
    ## Now we have to define the create_pin method.
        ## create_pin is a method of the AtmMachine class.
        ## Inside the create_pin method, we have to take the pin from the user and we have to take the balance from the user.
    ## Now we have to define the change_pin method.
        ## change_pin is a method of the AtmMachine class.
        ## Inside the change_pin method, we have to take the old pin from the user and we have to check the condition for the old pin.
    ## Now we have to define the check_balance method.
        ## check_balance is a method of the AtmMachine class.
        ## Inside the check_balance method, we have to take the pin from the user and we have to check the condition for the pin.
    ## Now we have to define the withdraw_balance method.
        ## withdraw_balance is a method of the AtmMachine class.
        ## Inside the withdraw_balance method, we have to take the pin from the user and we have to check the condition for the pin and we have to check the condition for the amount.
    
    
    ## Now we have to create the object of the AtmMachine class.
    ## After creating the object of the AtmMachine class, it will automatically call the constructor function.
    ## After calling the constructor function, it will automatically call the menu method.
    ## After calling the menu method, it will automatically call the create_pin method.
    ## After calling the create_pin method, it will automatically call the menu method.
    ## After calling the menu method, it will automatically call the change_pin method.
    ## After calling the change_pin method, it will automatically call the menu method.
    ## After calling the menu method, it will automatically call the check_balance method.
    ## After calling the check_balance method, it will automatically call the menu method.
    ## After calling the menu method, it will automatically call the withdraw_balance method.
    ## After calling the withdraw_balance method, it will automatically call the menu method.
    ## After calling the menu method, it will automatically call the exit function.
    ## After calling the exit function, it will exit the program.
    
    ## This is how the above code works.
    
    
    


````

````python


## For the ICICI ATM Machine we have to create a new class and we have to inherit the properties of the AtmMachine class.
class AtmMachineSBI:            # AtmMachine is a class name and where ATM and Machine are two words and M is capital letter of Machine and A is capital letter of ATM . 
    ## First of all we have to define data properties of Atm Machine inside the AtmMachine class.
        ## Data properties are the properties which are used to store data in the class.
        ## Data properties are also called as attributes of the class.
    ## Data properties are defined inside the class but outside the methods. 
    
    ## Q) How to define data properties in the class? 
        ## => To define data properties in the class,for this we need to use one special 
            # function inside a class which is called as Constructor. 
                ## Q) What is Constructor? 
                    ## => Constructor is a special function which is used to initialize the data properties of the class.
                    ## Whenever we see __init__ function inside a class, it means it is a constructor function.
                        # Special function has some super powers.
    ## Whatever data property we want to write , 
        # we have to write inside the constructor function.
    ## We need three data properties like pin number, balance and withdraw. 
    
    ## Constructor (special function) -> super power -> initialize the data properties of the class.
    
    def __init__(self):     # Constructor function is defined inside the class.
        self.pin = ""      # self is a keyword which is used to define the data properties of the class and it is used to access the data properties of the class.
        self.balance = 0   
        self.withdraw = 0  
        self.menu()        # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
        
    ## Now we have to define the methods of the class.
        ## Methods are the functions which are defined inside the class.
        ## Methods are used to perform some operations on the data properties of the class.
    
    ## We need to define 4 methods for the AtmMachine class. 
        ## 1) check_balance
        ## 2) withdraw 
        ## 3) pin number
        
    ## Now we have to define the menu method.
    def menu(self):                    # menu is a method of the AtmMachine class and self is a keyword which is used to access the data properties of the class.
        user_input = input(          # input is a function which is used to take input from the user and it is used to take input in the form of string.
            """
            Hi how can I help you ?
            1. Press 1 to create pin 
            2. Press 2 to change pin
            3. Press 3 to check balance
            4. Press 4 to withdraw money
            5. Anything to exit 
        
            """
        )
        
        ## Now we have to write the logic for the above options.
        if user_input == "1":          # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            ## create pin
            self.create_pin()          # self is a keyword which is used to access the data properties of the class and create_pin is a method of the AtmMachine class.
        
        elif user_input == "2":       # elif is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the elif
            ## change pin
            self.change_pin()          # self is a keyword which is used to access the data properties of the class and change_pin is a method of the AtmMachine class.
        
        elif user_input == "3":      # elif is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the elif
            ## check balance
            self.check_balance()      # self is a keyword which is used to access the data properties of the class and check_balance is a method of the AtmMachine class.
        
        elif user_input == "4":       # elif is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the elif
            ## withdraw money
            self.withdraw_balance()    # self is a keyword which is used to access the data properties of the class and withdraw_balance is a method of the AtmMachine class.
        
        else:                         # else is a keyword which is used to execute the block of code if the condition is false.
            exit()
            
    
    ## Now we have to define the create_pin method.
    def create_pin(self):              # create_pin is a method of the AtmMachine class and self is a keyword which is used to access the data properties of the class.
        ## Now we have to take the pin from the user.
        user_pin = input("Enter your pin: ")     # input is a function which is used to take input from the user and it is used to take input in the form of string.
        self.pin = user_pin
        
        ## Now we have to take the balance from the user.
        user_balance = int(input("Enter Balance:"))   # input is a function which is used to take input from the user and it is used to take input in the form of string. So we have to convert it into integer.
        self.balance = user_balance                 # self is a keyword which is used to define the data properties of the class and it is used to access the data properties of the class.
        
        print("Pin Created Successfully")          # print is a function which is used to print the message on the screen.
        self.menu()                               # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
    
    ## Now we have to define the change_pin method.
    def change_pin(self):
        old_pin = input("Enter your old pin: ")    # input is a function which is used to take input from the user and it is used to take input in the form of string.
        
        if old_pin == self.pin:                    # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            new_pin = input("Enter your new pin: ")   # input is a function which is used to take input from the user and it is used to take input in the form of string.
            self.pin = new_pin                        # self is a keyword which is used to access the data properties of the class.
            print("Pin Changed Successfully")         # print is a function which is used to print the message on the screen and it is used to print the message on the screen.
            self.menu()                              # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
        else:                                         # else is a keyword which is used to execute the block of code if the condition is false.
            print("Invalid Pin")                      # print is a function which is used to print the message on the screen.
            self.menu()                               # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
    
    ## Now we have to define the check_balance method.
    
    def check_balance(self):
        user_pin = input("Enter your pin: ")         # input is a function which is used to take input from the user and it is used to take input in the form of string.
        if user_pin == self.pin:                      # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            print(f"Your balance is {self.balance}")  # print is a function which is used to print the message on the screen and it is used to print the message on the screen.
        
        else:                                         # else is a keyword which is used to execute the block of code if the condition is false.
            print("Your pin is incorrect, please try again")            # print is a function which is used to print the message on the screen.
        self.menu()
    
    ## Now we have to define the withdraw method.
    
    def withdraw_balance(self):                       # withdraw_balance is a method of the AtmMachine class and self is a keyword which is used to access the data properties of the class.
        user_pin = input("Enter your pin: ")          # input is a function which is used to take input from the user and it is used to take input in the form of string.
        if user_pin == self.pin:                      # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            # allow to withdraw
            amount = int(input("Enter the amount : "))   # input is a function which is used to take input from the user and it is used to take input in the form of string. So we have to convert it into integer.
            if amount <= self.balance:                                       # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
                self.balance = self.balance - amount                         # self is a keyword which is used to access the data properties of the class and it is used to define the data properties of the class.
                print(f"You have withdrawn {amount}. Your new balance is {self.balance}")          # print is a function which is used to print the message on the screen.
            
            else:                                                            # else is a keyword which is used to execute the block of code if the condition is false.
                print("Insufficient balance!")                                # print is a function which is used to print the message on the screen.
        
        else:                                                                # else is a keyword which is used to execute the block of code if the condition is false and it is used to execute the block of code if the condition is false.
            print("Your pin is incorrect, please try again")                 # print is a function which is used to print the message on the screen.
        self.menu()                                                         # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
        
        


## For the SBI ATM Machine, we have to create the object of the AtmMachineSBI class and call the menu method using the object of the class to run the program.
class AtmMachineICICI:            # AtmMachine is a class name and where ATM and Machine are two words and M is capital letter of Machine and A is capital letter of ATM . 
    ## First of all we have to define data properties of Atm Machine inside the AtmMachine class.
        ## Data properties are the properties which are used to store data in the class.
        ## Data properties are also called as attributes of the class.
    ## Data properties are defined inside the class but outside the methods. 
    
    ## Q) How to define data properties in the class? 
        ## => To define data properties in the class,for this we need to use one special 
            # function inside a class which is called as Constructor. 
                ## Q) What is Constructor? 
                    ## => Constructor is a special function which is used to initialize the data properties of the class.
                    ## Whenever we see __init__ function inside a class, it means it is a constructor function.
                        # Special function has some super powers.
    ## Whatever data property we want to write , 
        # we have to write inside the constructor function.
    ## We need three data properties like pin number, balance and withdraw. 
    
    ## Constructor (special function) -> super power -> initialize the data properties of the class.
    
    def __init__(self):     # Constructor function is defined inside the class.
        self.pin = ""      # self is a keyword which is used to define the data properties of the class and it is used to access the data properties of the class.
        self.balance = 0   
        self.withdraw = 0  
        self.menu()        # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
        
    ## Now we have to define the methods of the class.
        ## Methods are the functions which are defined inside the class.
        ## Methods are used to perform some operations on the data properties of the class.
    
    ## We need to define 4 methods for the AtmMachine class. 
        ## 1) check_balance
        ## 2) withdraw 
        ## 3) pin number
        
    ## Now we have to define the menu method.
    def menu(self):                    # menu is a method of the AtmMachine class and self is a keyword which is used to access the data properties of the class.
        user_input = input(          # input is a function which is used to take input from the user and it is used to take input in the form of string.
            """
            Hi how can I help you ?
            1. Press 1 to create pin 
            2. Press 2 to change pin
            3. Press 3 to check balance
            4. Press 4 to withdraw money
            5. Anything to exit 
        
            """
        )
        
        ## Now we have to write the logic for the above options.
        if user_input == "1":          # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            ## create pin
            self.create_pin()          # self is a keyword which is used to access the data properties of the class and create_pin is a method of the AtmMachine class.
        
        elif user_input == "2":       # elif is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the elif
            ## change pin
            self.change_pin()          # self is a keyword which is used to access the data properties of the class and change_pin is a method of the AtmMachine class.
        
        elif user_input == "3":      # elif is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the elif
            ## check balance
            self.check_balance()      # self is a keyword which is used to access the data properties of the class and check_balance is a method of the AtmMachine class.
        
        elif user_input == "4":       # elif is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the elif
            ## withdraw money
            self.withdraw_balance()    # self is a keyword which is used to access the data properties of the class and withdraw_balance is a method of the AtmMachine class.
        
        else:                         # else is a keyword which is used to execute the block of code if the condition is false.
            exit()
            
    
    ## Now we have to define the create_pin method.
    def create_pin(self):              # create_pin is a method of the AtmMachine class and self is a keyword which is used to access the data properties of the class.
        ## Now we have to take the pin from the user.
        user_pin = input("Enter your pin: ")     # input is a function which is used to take input from the user and it is used to take input in the form of string.
        self.pin = user_pin
        
        ## Now we have to take the balance from the user.
        user_balance = int(input("Enter Balance:"))   # input is a function which is used to take input from the user and it is used to take input in the form of string. So we have to convert it into integer.
        self.balance = user_balance                 # self is a keyword which is used to define the data properties of the class and it is used to access the data properties of the class.
        
        print("Pin Created Successfully")          # print is a function which is used to print the message on the screen.
        self.menu()                               # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
    
    ## Now we have to define the change_pin method.
    def change_pin(self):
        old_pin = input("Enter your old pin: ")    # input is a function which is used to take input from the user and it is used to take input in the form of string.
        
        if old_pin == self.pin:                    # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            new_pin = input("Enter your new pin: ")   # input is a function which is used to take input from the user and it is used to take input in the form of string.
            self.pin = new_pin                        # self is a keyword which is used to access the data properties of the class.
            print("Pin Changed Successfully")         # print is a function which is used to print the message on the screen and it is used to print the message on the screen.
            self.menu()                              # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
        else:                                         # else is a keyword which is used to execute the block of code if the condition is false.
            print("Invalid Pin")                      # print is a function which is used to print the message on the screen.
            self.menu()                               # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
    
    ## Now we have to define the check_balance method.
    
    def check_balance(self):
        user_pin = input("Enter your pin: ")         # input is a function which is used to take input from the user and it is used to take input in the form of string.
        if user_pin == self.pin:                      # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            print(f"Your balance is {self.balance}")  # print is a function which is used to print the message on the screen and it is used to print the message on the screen.
        
        else:                                         # else is a keyword which is used to execute the block of code if the condition is false.
            print("Your pin is incorrect, please try again")            # print is a function which is used to print the message on the screen.
        self.menu()
    
    ## Now we have to define the withdraw method.
    
    def withdraw_balance(self):                       # withdraw_balance is a method of the AtmMachine class and self is a keyword which is used to access the data properties of the class.
        user_pin = input("Enter your pin: ")          # input is a function which is used to take input from the user and it is used to take input in the form of string.
        if user_pin == self.pin:                      # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
            # allow to withdraw
            amount = int(input("Enter the amount : "))   # input is a function which is used to take input from the user and it is used to take input in the form of string. So we have to convert it into integer.
            if amount <= self.balance:                                       # if is a keyword which is used to check the condition and if the condition is true then it will execute the block of code inside the if block.
                self.balance = self.balance - amount                         # self is a keyword which is used to access the data properties of the class and it is used to define the data properties of the class.
                print(f"You have withdrawn {amount}. Your new balance is {self.balance}")          # print is a function which is used to print the message on the screen.
            
            else:                                                            # else is a keyword which is used to execute the block of code if the condition is false.
                print("Insufficient balance!")                                # print is a function which is used to print the message on the screen.
        
        else:                                                                # else is a keyword which is used to execute the block of code if the condition is false and it is used to execute the block of code if the condition is false.
            print("Your pin is incorrect, please try again")                 # print is a function which is used to print the message on the screen.
        self.menu()                                                         # self is a keyword which is used to access the data properties of the class and menu is a method of the AtmMachine class.
        


````
