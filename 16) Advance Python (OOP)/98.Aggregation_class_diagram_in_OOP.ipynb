# Aggregation Class Diagram in OOP :

Aggregation is a special form of association. It is a relationship between two classes like association, however its a directional association, which means it is strictly a one way relationship. It represents a "whole-part" relationship. In aggregation, the child can exist independently of the parent. So, it is a weak relationship.

If you want to show aggregation in class diagram, you can use a **diamond shape** at the parent class and a line with an arrow at the child class.
**Whoever having this particular Diamond shape, that means this particular class owns the other class.** This is called aggregation diagram in OOP.

In the parent class block, There are three small blocks ,
1. First block contain the name of the class which is customer class.
2. Second block contain the name of the attributes or variables which are name, gender, address. these all are public properties because I'm not using any private properties here. So, I'm going to give the plus sign in front of the attributes.
