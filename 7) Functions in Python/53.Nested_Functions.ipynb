# Nested Functions in Python :

In previous we have learn about nested for loops , nested while loops and nested if-else statements. 

Nested for loop mean a loop inside a loop. Similarly, nested if-else mean if-else inside if-else. 

**In the same way, nested functions mean a function inside a function.**

Nested functions are normally used to hide the functionality inside a function ,  So that the other person can't access the hidden functionality of the function. 
It will make the code more readable and understandable.


```python
## Q) Write a Program to demonstrate nested functions in Python ?

def f():   # Outer function
    def g():   # Inner function
        print("Inside function g")  # Printing message
    
    g()  # Calling inner function
    print("Inside function f") # Printing message

f() # Calling outer function

# First f() function is called, which in turn calls g() function 
    # and g() function prints the message "Inside function g" and then f() function prints the message "Inside function f". 
    # This is how nested functions work in Python.

# We can also call the inner function g() from outside the outer function f() as shown below.
    # f().g() # This will give an error as g() is not defined outside f() function.

# But we can't call the inner function g() directly it will give an error as shown below.
    # g() # This will give an error as g() is not defined outside f() function. 
```

```python
## Q) Write a Program to demonstrate infinite recursion using nested functions in Python ?

# This will result in infinite recursion as f() calls g() and g() calls f() and so on.

def f():  # Outer function
    def g(): # Inner function
        print("Inside function g")
        f()  # Calling outer function f() from inner function g() 
                # This will result in infinite recursion as f() calls g() and g() calls f() and so on.
    
    g() # Calling inner function
    print("Inside function f")

f()  # Calling outer function

# Output:
# Inside function g  # This will be printed multiple times 

# This will continue until the maximum recursion depth is reached and 
    # So, it will throw an error "RecursionError: maximum recursion depth exceeded in comparison".
```
