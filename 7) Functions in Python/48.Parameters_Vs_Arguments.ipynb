# Parameters Vs Arguments : 

## Parameters : 
- Parameters are the variables that are defined in the function definition.
- They are used to pass the values to the function.
- They are defined in the function definition.

## Arguments :
- Arguments are the values that are passed to the function.
- They are used to pass the values to the function.
- They are defined in the function call.



````python

## This is a simple python script to check if a given number is odd or even.

# Parameters in a function are the variables that are used in the function definition.
def odd_even(num): ## Parameters 
    """
    This function checks if a given number is odd or even.
    input: any valid number/integer
    output: odd / even
    Created Date: 8-14-2024
    """
    if type(num) == int:
        if num % 2 == 0:
            return "Even Number"
        else:
            return "Odd Number"
    else:
        return "Wrong input"

# Arguments in a function are the values that are passed to the function when it is called. 
res = odd_even(2) ## Arguments 
print(res)
````

# Types of Arguments :

- Default Argument
- Positional Argument
- Keyword Argument

##### Default Argument :
- Default arguments are the arguments that take the default value if no value is passed to the function.
- The default value is assigned to the parameter in the function definition.
- The default value is assigned using the assignment operator(=). 

We use default arguments when we want to assign a default value to the parameter if no value is passed to the function.

Syntax : 
```python
def function_name(parameter_name=default_value):
    # code
    return
```

Example : 
```python
def power(a=2,b=3): # default values are assigned to a and b 
    return a**b 

power() # 8  # default values are used
power(2,3) # 8  # values are passed to the function but it was same as default values so it will return the same output

```

##### Positional Argument :
- Positional arguments are the arguments that are passed to the function based on the position of the parameter in the function definition.

If we know the position of the parameter in the function definition then we can use positional arguments.

Syntax : 
```python
def function_name(parameter1,parameter2):
    # code
    return
```

Example : 
```python
def power(a,b):  # a is the first parameter and b is the second parameter 
    return a**b 

# values are passed based on the position of the parameter in the function definition
power(2,3) # 8  # 2 is passed to a and 3 is passed to b
power(3,2) # 9  # 3 is passed to a and 2 is passed to b
```

##### Keyword Argument :
- Keyword arguments are the arguments that are passed to the function based on the parameter name in the function definition.

If we don't know the position of  a and b because there are many parameters in the function definition , it is not possible to remember the all the positions of the parameters in the function definition so we can use keyword arguments.

Keywords arguments are more readable than positional arguments because we can pass the values to the function based on the parameter name in the function definition.

In project development, we use keyword arguments because we don't know the position of the parameter in the function definition. We can pass the values to the function based on the parameter name in the function definition.


Syntax : 
```python
def function_name(parameter1,parameter2):
    # code
    return
```

Example : 
```python
def power(a=2,b=3):
    return a**b

# values are passed based on the parameter name in the function definition
power(b=2,a=3) # 9  # 3 is passed to a and 2 is passed to b 

```
