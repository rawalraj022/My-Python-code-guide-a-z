### What are functions in Python? 

**Functions is a Programming construct that is used to define a block of code that can be called . It takes inputs & returns outputs. Inputs are called arguments & outputs are called return values.**

Functions are used to perform a specific task. Functions are used to break down large programs into smaller and modular programs. As our program grows larger and larger, functions make it more organized and manageable. Furthermore, it avoids repetition and makes the code reusable. 

Until now, we have been using Python's built-in functions like print(), input(), len(), etc. But we can also create our own functions, called user-defined functions.

**There are 2 types of functions in Python:**

**1. Built-in functions**    
[ print(), input(), len(), int(), float(), str(), list(), dict(), tuple(), set() ]
- Built-in functions are those functions that are already defined in Python. We can use them directly. We have already learned Built-in functions in Python. 

**2. User-defined functions** 

- User-defined functions are those functions that we define ourselves. We can define the function once and use it multiple times. 

Let's say if I'm talking about **Facebook** , Facebook having different different function, We can create our own custom logging function, We can create our own custom signin function , We can create our own custom profile creater function , We can create our own custom posting function . So they combined all the unique function together and they created a big software called Facebook application.

#### Advantages of Functions in Python :

1. **Code Reusability** : We can define the function once and use it multiple times. By calling the function, we can use the same code multiple times. By this, we can avoid writing the same code again and again. By this, we can save time and effort. 
Example: If we want to add two numbers multiple times, we can define the function once and use it multiple times like add(2,3), add(4,5), add(6,7), etc. and function code is defined only once. 

2. **Modularity** : We can break down the large program into smaller and manageable pieces. Each piece of code will perform a specific task. By this, the code becomes more organized and manageable. Each function will perform a specific task.

3. **Easy Debugging** : If there is an issue in the code, we can debug the code easily. We can debug the code function by function. We can identify the issue in the specific function and fix it. 
Debugging becomes easier because if there is an issue in login function, we can debug the login function only. We don't need to debug the entire code.

4. **Readability** : The code becomes more readable and understandable. We can understand the code function by function. Each function will perform a specific task. The code becomes more organized and readable. 
It makes the different developers to work on the same code because the code is more organized and readable and understandable.

5. **Less Coding** : We can avoid writing the same code again and again. We can define the function once and use it multiple times. By this, we can save time and effort. We can avoid writing the same code again and again.

# Creating a Function in Python : 

In Python, we can create a function using the def keyword. The syntax of creating a function is as follows :

Syntax : 
```python
def function_name(arguments):
    # code
    return value
```

- **def** : def is a keyword in python. It is used to define the function.

- **function_name** : function_name is the name of the function. We can give any name to the function. The function name should be meaningful.

- **arguments** : arguments are the inputs to the function. We can pass the arguments to the function. Arguments are the values that we pass to the function.

- **code** : code is the logic of the function. It is a block of code that performs a specific task. The code is defined inside the function.

- **return** : return is a keyword in python. It is used to return the value from the function. The return value is the output of the function. We can return the value from the function.

Example : 
```python
def add(a,b):
    c = a+b
    return c
```

In the above example, we have created a function called add. The function name is add. The function takes two arguments a and b. The function returns the sum of a and b. The function returns the value c. The function returns the sum of a and b.

````python

"""
    Writing Simple script to print different names of the people ?
    

    ```python
    Print("Hello Rajkumar, Welcome to Python")
    Print("Hello Ram, Welcome to Python")
    Print("Hello Sita, Welcome to Python")
    ``` 

    Output:

    Hello Rajkumar, Welcome to Python
    Hello Ram, Welcome to Python
    Hello Sita, Welcome to Python
    

"""
# => If we have to print the names of 100 people, then we have to write 100 print 
    # statements which is not a good practice to write the code in this way.

# Solution: We create a function which takes the name as an argument and print the name. 

def greet(name):   # Here name is the argument which is passed to the function greet.
    output = f"Hello {name}, Welcome to Python"  # Here we are using f-string to print the name of the person and welcome message.
    return(output)  # Here we are returning the output.

# Now we can call the function greet with the name of the person as an argument to print the name of the person.
res = greet("Rajkumar")
print(res)
````

````python
## Q) If there are thousands names of the people , we want to show the greeting message to all the people ?   

# Creating the function greet with the parameter name to print the name of the person and welcome message.
def greet(name):
    output = f"Hello {name}, Welcome to Python"
    return output

# Creating the list of names of the people.
names = ["Rajkumar", "Ram", "Sita", "Laxman", "Bharat", "Shatrughan", "Hanuman", "Vibhishan", "Kumbhkaran", "Meghnath"] 

# Using for loop to iterate over the list of names and print the name of the person and welcome message.

for name in names:  # Here we are iterating over the list of names using for loop and storing the name of the person in the variable name.
    res = greet(name) # Here we are calling the function greet with the name of the person as an argument and storing the output in the variable res.
    print(res)  # Here we are printing the name of the person and welcome message on the console.

# Defining the function greet one time and using it multiple times to print the name of the persons and welcome message to all the people.
    # In this way, we can avoid writing the same code multiple times. 
````

````python
## Q) Checking if the number is odd or even number and print the number is odd or even number using python function ?

# Creating the function odd_even with the parameter num to check if the number is odd or even number.

def odd_even(num):  # Here num is the argument which is passed to the function odd_even. 'def' is the keyword used to define the function.
    if num % 2 == 0:  # Here we are using if condition to check if the number is divisible by 2 and the remainder is 0 or not. % is the modulus operator which is used to find the remainder.
        return "Even number"  # Here we are returning the output if the number is even number.
    else :                     # Here we are using else condition if the number is not divisible by 2 and the remainder is not 0.
        return "Odd number"

# Using for loop to iterate over the range of numbers from 1 to 10 and print the number and whether the number is odd or even number.
    # Calling the function odd_even with the number as an argument to check if the number is odd or even number.

for i in range(1,11):  # Here we are using for loop to iterate over the range of numbers from 1 to 10.
    res = odd_even(i)  # Here we are calling the function odd_even with the number as an argument and storing the output in the variable res.
    print(f"{i} is {res}")  # Here we are printing the number and whether the number is odd or even number on the console.


# Defining the function odd_even one time and using it multiple times to check if the number is odd or even number and print the number is odd or even number.
    # In this way, we can avoid writing the same code multiple times.
````
