#### 1) Code Modularity:

Code modularity means dividing the code into separate modules. 
**Modules are a set of related functions mean each module is composed of a set of related functions.**

- It is the process of breaking a computer program into separate modules. 
- Each module is composed of a set of related functions. 
- These functions are called by the main program whenever needed. 
- This makes the code easier to understand and maintain.

**Example :** Like in facebook login page we have one module for login, one module for signup, one module for forgot password. In each module, like in signup module we have functions like validate email, validate password, validate phone number, validate name, etc. So, Signup module is contain of validate email function, validate password function, validate phone number function, validate name function. So, this is called code modularity. 

In Forgot password module we have functions like validate email, send otp, validate otp, reset password, etc. So, Forgot password module is contain of validate email function, send otp function, validate otp function, reset password function. So, this is called code modularity.

In login module we have functions like validate email, validate password, login, etc. So, login module is contain of validate email function, validate password function, login function. So, this is called code modularity.

So, code modularity means dividing the code into separate modules. Modules are a set of related functions. These functions are called by the main program whenever needed. This makes the code easier to understand and maintain.
