# Enumerate Function in Python :

Enumerate function is very helpful while working with the index of the elements in the list. It is used to loop through the containers like list, tuple, string, etc. It returns the index and the value of all the items in the container as a tuple.


**Syntax of Enumerate function :**

```python
enumerate(iterable, start=0)
```

**Parameters :**

1. **iterable** : The iterable to be looped through.
2. **start** : The index from which the index should start. By default, it is 0.

**Return Value :**

It returns an enumerate object that contains the index and the value of all the items in the container as a tuple.

#### Using Enumerate function :

```python
# List of strings
L = ["Rajkumar", "Rahul", "Rohit", "Ravi"]

for index, name in enumerate(L):
    print(index, name)
```

Output :

```
0 Rajkumar
1 Rahul
2 Rohit
3 Ravi
```

Explanation of the above code : 

1. The enumerate function takes the list as an argument and returns the index and the value of the elements in the list.
2. The for loop iterates through the list and prints the index and the value of the elements in the list.



````python
## Q) If we want to start the index at 1, we can use the enumerate function to get the index and the value of the list ? 

L = ["Raj", "Rahul", "Ravi", "Rahim", "Ramesh"]

for index, name in enumerate(L, start=1):   # start=1 will start the index at 1 
                                            # by default it starts at 0
    print(index, name)

````
