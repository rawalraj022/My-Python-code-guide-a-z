# Operations on Lists 

Operation means to perform some operation on the list. The operation can be anything like adding elements, removing elements, updating elements, etc.

Some of the operations we can perform on lists are: 

1) Arithmetic Operations
2) Membership Operations
3) Looping Operations 


### 1) Arithmetic Operations

Arithmetic operations are the operations that we perform on the list elements. 

Inside the arithmetic operations we can only perform addition '+' and multiplication '*' operations. Like adding two lists, multiplying the list with a number, etc.

#### 1.1) Adding two lists

We can add two lists by using the '+' operator. The '+' operator concatenates the two lists and returns a new list.

```python

L1 = [1, 2, 3]
L2 = [4, 5, 6]

Sum = L1 + L2  # Doing concatenation of two lists , Concatenation means adding two lists .
print(Sum) 

```

Output:
```
[1, 2, 3, 4, 5, 6]
```

#### 1.2) Multiplying the list with a number

We can multiply the list with a number by using the '*' operator. The '*' operator repeats the list elements by the given number of times.

We can't multiply two lists directly because it is not supported in python.

Multiplying the list with 3 , It will repeat the list elements 3 times. But it will not change the original list. and return a new list.

```python   

L = [1, 2, 3]

Mul = L * 3  # Multiplying the list with 3 , It will repeat the list elements 3 times. But it will not change the original list. and return a new list. and We can't multiply two lists directly because it is not supported in python.

print(Mul)

```

Output:
```
[1, 2, 3, 1, 2, 3, 1, 2, 3]
```


### 2) Membership Operations

Membership operations are the operations that we perform to check whether the element is present in the list or not or to check whether the list is a subset of another list or not. 

Inside the membership operations, we can use the 'in' and 'not in' operators.

#### 2.1) Check whether the element is present in the list or not

We can check whether the element is present in the list or not by using the 'in' operator. The 'in' operator returns True if the element is present in the list otherwise it returns False.

```python   

L = [1, 2, 3, 4, 5]

print(3 in L)  # It will return True because 3 is present in the list.

print(6 in L)  # It will return False because 6 is not present in the list.

```

Output:
```
True
False
```

#### 2.2) Check whether the list is a subset of another list or not

We can check whether the list is a subset of another list or not by using the 'in' operator. The 'in' operator returns True if the list is a subset of another list otherwise it returns False.

```python

L1 = [1, 2, 3, 4, 5]
L2 = [2, 3]

print(L2 in L1)  # It will return True because L2 is a subset of L1.

L3 = [6, 7]

print(L3 in L1)  # It will return False because L3 is not a subset of L1.

```

Output:
```
True
False
```



### 3) Looping Operations 

Looping operations are the operations that we perform to iterate over the list elements. 

Inside the looping operations, we can use the 'for' loop to iterate over the list elements.

#### 3.1) Iterate over the list elements

We can iterate over the list elements by using the 'for' loop. The 'for' loop iterates over the list elements one by one.

```python

L = [1, 2, 3, 4, 5]

for i in L:  # Iterating over the list elements one by one.
    print(i)

```

Output:
```
1
2
3
4
5
```

#### 3.2) Iterate over the list elements with index

We can iterate over the list elements with index by using the 'enumerate' function. The 'enumerate' function returns the index and element of the list.

```python

L = [1, 2, 3, 4, 5]

for i, val in enumerate(L):  # Iterating over the list elements with index.
    print(i, val)

```

Output:
```
0 1 
1 2
2 3
3 4
4 5
```
