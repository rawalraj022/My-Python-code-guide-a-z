# Dictionary Operations in Python :

In Dictionary , we can only perform two operations . They are :

1. Membership Operations
2. Iteration Operations

## 1. Membership Operations :

In Membership Operations , we can check whether a key is present in the dictionary or not . We can use the following operators to check the membership of a key in the dictionary :

1. in
2. not in

### Example :

```python
# Python code to demonstrate
# checking of key in dictionary

# Initialising dictionary
test_dict = { "geeks" : 1, "for" : 2, "geek" : 3 }

# using in to check if key is present
print("Checking if 'geeks' in dictionary : ")
print("geeks" in test_dict)

# using in to check if key is present
print("Checking if 'for' in dictionary : ")
print("for" in test_dict)
```

### Output :

```
Checking if 'geeks' in dictionary :
True
Checking if 'for' in dictionary :
True
```

## 2. Iteration Operations :

In Iteration Operations , we can iterate through the dictionary using the following methods :

Example :

```python

# Python code to demonstrate
# iteration over dictionary

# Initialising dictionary
test_dict = { "geeks" : 1, "for" : 2, "geek" : 3 }

# using for loop to iterate
print("Iterating over dictionary : ")

for i in test_dict:
    print(i, test_dict[i])
```

### Output :

```
Iterating over dictionary :
geeks 1
for 2
geek 3
```
