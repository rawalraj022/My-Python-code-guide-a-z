# Dictionary Comprehension in Python : 

Dictionary comprehension is defined as a way to construct a dictionary by performing an operation on each item in an iterable. It is a concise way to create dictionaries. It will make your code more readable and clean. It is a more efficient way to create dictionaries.



```python
# Q) print 1st 10 numbers and their squares root ?

{i:i**2 for i in range(1,11)}  # i is key and i**2 is value in dictionary comprehension 
```

```python
## Using if condition in dictionary comprehension
# Q) write a program to print the key which is not having zero value in dictionary ?

products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

{ key:value for (key,value) in products.items() if value != 0}  # key is not having zero value in dictionary comprehension and it will print only key which is not having zero value
                                                                # key:value is tuple unpacking in dictionary comprehension
                                                                # items() is used to get key and value from dictionary
                                                                # if value != 0 is used to check the value is not zero in dictionary comprehension
                                                                # So, it will print only key which is not having zero value in dictionary comprehension
```

```python
## using zip
# Q) write a program to print days and temperature in side by side in dictionary comprehension ?

days = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

{ i:j for (i,j) in zip(days,temp_C)} # i is key and j is value in dictionary comprehension
                                        # zip is used to combine two lists into dictionary comprehension
                                        # it will print days as key and temp_C as value in dictionary comprehension
                                    
```
