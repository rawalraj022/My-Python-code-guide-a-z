# Problems with working in text mode 

So far, we have been working with textual data. Let's say if we have text file so how we can read it with the help of this particular open statement  => open('filename', 'r') . we can perform the read operation , write operation and append operation on the text file. 

But what about others files like image files, audio files, video files etc. 
So, we can't perform the read operation on these files with the help of this open statement => open('filename', 'r') because this open statement is only for the text files not for the binary files like image files, audio files, video files etc. 

So, we have to use the binary mode to read these files with the help of this open statement => open('filename', 'rb')

- It can't work with binary files like images
- not good for other data types like int/float/list/tuples



```python
## Working with binary file data

with open("files/screanshot.jpg", "rb") as f:
    print(f.read())


## Explanation of the above code
# The "rb" mode is used to read binary data from a file. The "r" stands for read and the "b" stands for binary.
# The "with" statement is used to open the file and automatically close it when the block of code is done executing.
# The "f" variable is used to reference the file object.
# The "read()" method is used to read the contents of the file and return it as a bytes object.
# The "print()" function is used to print the contents of the file to the console.
# The output will be a series of bytes that represent the binary data of the file.
# This is useful for working with binary files, such as images, audio files, and other non-text files.
```

```python
## working with binary file

# Copying a binary file
# You can use the "rb" and "wb" modes to read and write binary data to and from a file.
# This can be useful for copying binary files, such as images, audio files, and other non-text files.
# Here's an example of copying a binary file:

with open("files/screanshot.jpg", 'rb') as f:
    with open("files/screanshot_copy.jpg", 'wb') as wf:
        wf.write(f.read())


## Explanation of the above code
# The first "with" statement is used to open the source file in read binary mode ("rb").
# The "f" variable is used to reference the file object for the source file.    
# The second "with" statement is used to open the destination file in write binary mode ("wb").
# The "wf" variable is used to reference the file object for the destination file.
# The "read()" method is used to read the contents of the source file and return it as a bytes object.
# The "write()" method is used to write the contents of the source file to the destination file.
# This will copy the binary data from the source file to the destination file.
# The destination file will now contain an exact copy of the source file.
# This can be useful for creating backups of binary files or transferring them between systems.

```

```python
## working with integer data in binary files :

# In addition to reading and writing binary data, you can also work with other data types in binary files 
# by using the "struct" module in Python. The "struct" module provides functions for converting between 
# Python values and C structs represented as bytes objects.

with open("files/sample.txt", "w") as f:
    f.write(5)
```

```python
## working with integer data in binary files :

with open('files/sample1.txt','w') as f:
    f.write("5")

# The above code will write the string "5" to the file "sample1.txt".
# If you want to write the integer 5 to the file, you can convert it to a string first:
```

```python
## working with dictionary data types in binary files : 

d = {
    'name':'Raj',
    'age':23,
    'gender':'male'
}

with open('files/sample1.txt','w') as f:
    f.write(d)

# The above code will raise an error because you cannot write a dictionary directly to a file
# You can convert the dictionary to a string first using the "str()" function:
# The "str()" function converts the dictionary to a string representation that can be written to a file.
```

```python
## working with dictionary data types in binary files :

d = {
    'name':'Bappy',
    'age':23,
    'gender':'male'
}

with open('files/sample1.txt','w') as f:
    f.write(str(d))

# The above code will write the string representation of the dictionary to the file "sample1.txt".


```

```python
## If we want to read the string representation of the dictionary from the file :

with open('files/sample1.txt','r') as f:
    s = f.read()
    print(s)
    print(type(s))

# The "read()" method is used to read the contents of the file and return it as a string.
```

```python
## If we want to convert the string representation of the dictionary back to a dictionary :

with open('files/sample1.txt','r') as f:
    s = f.read()
    print(dict(s))

# The above code will raise an error because you cannot convert a string to a dictionary directly using the "dict()" function.
```
