# Common Functions/Methods in Python :

Before starting with our inbuilt method inside the string first of all we want to discuss some of the common function/methods which are used in python. 

Common function/method actually we'll be using in every datatype not only in string but it can also support in other data types as well like list, tuple, dictionary, etc. 

So, let's start with some of the common methods/functions in python.

#### 1. **len()** : 
This function is used to find the length of the string. It can also be used to find the length of the list, tuple, dictionary, etc. data types.

let's see an example of how we can use this function in python.

```python
string = "Hello, World!"
print(len(string))
```

Output : 13

#### 2. **max()** : 
This function is used to find the maximum value from the string. It can also be used to find the maximum value from the list, tuple, dictionary, etc. data types.

let's see an example of how we can use this function in python.

```python
string = "Hello, World!" # In this string the maximum value is r , how r is the maximum value because in the ASCII value of r is 114 which is greater than all the other values.

print(max(string))  # max() function will return the maximum value from the string.
```

Output : r   

#### 3. **min()** : 
This function is used to find the minimum value from the string. It can also be used to find the minimum value from the list, tuple, dictionary, etc. data types. 

let's see an example of how we can use this function in python.

```python
string = "Hello, World!" # In this string the minimum value is space, how space is the minimum value because in the ASCII value of space is 32 which is less than all the other values.

print(min(string))  # min() function will return the minimum value from the string.
```

Output :    

#### 4. **sorted()** : 
This function is used to sort the string in ascending order. It can also be used to sort the list, tuple, dictionary, etc. data types.

let's see an example of how we can use this function in python.

```python   
string = "Hello, World!"

print(sorted(string))  # sorted() function will return the sorted string in ascending order.
```

Output : [' ', '!', ',', 'H', 'W', 'd', 'e', 'l', 'l', 'l', 'o', 'o', 'r']

reverse = True : This parameter is used to sort the string in descending order.

let's see an example of how we can use this parameter in python.

```python   
string = "Hello, World!"

print(sorted(string, reverse = True))  # sorted() function will return the sorted string in descending order.
```

Output : ['r', 'o', 'o', 'l', 'l', 'l', 'e', 'd', 'W', 'H', ',', '!', ' ']

#### 1. **capitalize()** :
This method is used to convert the first character of the string to uppercase and remaining characters to lowercase.

let's see an example of how we can use this method in python.

```python   
string = "hello, world!"

print(string.capitalize())  # capitalize() method will return the first character of the string to uppercase and remaining characters to lowercase.
```

Output : Hello, world!


````python
s = "welcome to python"
print(s.capitalize())  # It will capitalize the first letter of the string and make all other letters small 
````

#### 11. **format()** :

This method is used to format the string in a specified format by using curly braces {} as placeholders for the variables which will be passed to the format() method as arguments. 

let's see an example of how we can use this method in python.

```python
string = "Hello, {}!"

print(string.format('World'))  # format() method will format the string in a specified format by using curly braces {} as placeholders for the variables which will be passed to the format() method as arguments.
```

Output : Hello, World!


````python
Name = "Rajkumar"
gender = "Male"

s = "My name is {0} and I am a {1}"  # It will replace the {0} with the value of the variable Name and {1} with the value of the variable 
                                    # 0 and 1 are the index of the variables

print(s.format(Name, gender)) # It will replace the {} with the values of the variables|
````

````python
# Facebook login success message print using format method for personalization name 

s = "Welcome {} to Facebook"

name = "Rajkumar"

print(s.format(name)) # It will replace the {} with the value of the variable name 
````

#### 15. **isidentifier()** :

This method is used to check whether the string is a valid identifier or not like variable name, function name, etc.

- We can't use the reserved keywords as an identifier like if, else, for, etc. 

- We can't use the special characters as an identifier like @, #, $, etc.

- We can't start the identifier with a number but we can use the number in between the identifier.

- We can't use the space in between the identifier but we can use the underscore _ in between the identifier like my_name, my_age, etc.

let's see an example of how we can use this method in python.

```python
string = "Hello, World!" # In this string we are having space in between the identifier so it will return False.

print(string.isidentifier())  # isidentifier() method will return True if the string is a valid identifier otherwise it will return False.
```

Output : False


````python
s = "name_age"

print(s.isidentifier()) # It will return True if the string is a valid identifier else it will return False
````

#### 16. **split()** :

This method is used to split the string into a list of substrings based on the specified delimiter. 

List is data type in python which is collection of items which are ordered and changeable.

let's see an example of how we can use this method in python.

```python
string = "Hello, World!"  # In this string we are having comma as a delimiter so it will split the string into a list of substrings based on the comma. 

print(string.split(','))  # split() method will split the string into a list of substrings based on the specified delimiter and return the list of substrings.
```

Output : ['Hello', ' World!'] 



````python

s = "My name is rajkumar"  # It will split the string into words on the basis of space and return a list of words 

print(s.split()) # It will split the string into words and return a list of words 
````

````python
s = "My,name,is,rajkumar" # In this string we are having comma as a delimiter so it will split the string into a list of substrings based on the comma.

print(s.split(','))  # split() method will split the string into a list of substrings based on the specified delimiter and return the list of substrings.
````

````python
s = "My@name@is@rajkumar" # In this string we are having @ as a delimiter so it will split the string into a list of substrings based on the @.

print(s.split('@'))  # split() method will split the string into a list of substrings based on the specified delimiter and return the list of substrings.
````

#### 17. **join()** :

This method is used to join the elements of the list into a string based on the specified delimiter.

Like split() method is used to split the string into a list of substrings based on the specified delimiter whereas join() method is used to join the elements of the list into a string based on the specified delimiter.

let's see an example of how we can use this method in python.

```python
string = ['Hello', 'World!']  # In this list we are having comma as a delimiter so it will join the elements of the list into a string based on the comma.

print(','.join(string))  # join() method will join the elements of the list into a string based on the specified delimiter and return the string.
```

Output : Hello,World!


````python

words = ['My', 'name', 'is', 'Rajkumar']

s = " ".join(words) # It will join the list of words into a string with space as a separator

print(s) # It will join the list of words into a string with space as a separator
````

````python
words = ['My', 'name', 'is', 'Rajkumar']

s = "_".join(words) # It will join the list of words into a string with space as a separator

print(s) # It will join the list of words into a string with space as a separator
````

#### 18. **replace()** :

#### 19. **strip()** :

This method is used to remove the leading and trailing whitespaces from the given string.

Eg : In facebook when we are entering the password if we are entering the password with leading and trailing whitespaces then it will remove the whitespaces and then it will check the password.

let's see an example of how we can use this method in python.

```python
string = " Hello, World! "  # In this string we are having leading and trailing whitespaces so it will remove the leading and trailing whitespaces.

print(string.strip())  # strip() method will remove the leading and trailing whitespaces from the given string and return the string.
```

Output : Hello, World!


````python
# without using strip() method 

s = "My name is Rajkumar   " 
print(s)
len(s) # It will return the length of the string
````

````python
# using strip() method

s = "My name is Rajkumar   "
s = s.strip() # It will remove the leading and trailing whitespaces from the string
print(s)
len(s) # It will return the length of the string
````
