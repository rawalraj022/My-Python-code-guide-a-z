# # Variables :

Variable is a container and it can store any kinds of value ( datatype ) .

### **Examples of variables :**



```python
a = 2 # Here 'a' is a variable and 2 is a value of the variable 'a' and it's an integer datatype.  

print(a)  # Here we're printing the value of the variable 'a' .

# Previously we are directly passing the data in print function, but now we're passing the variable in print function. 
```

```python
b = 3.5  # Here 'b' is a variable and 3.5 is a value of the variable 'b' and it's a float datatype.

print(b)  # Here we're printing the value of the variable 'b' .
```

```python
# We can also print multiple variable in one particular print function .

c = True # Here 'c' is a variable and True is a value of the variable 'c' and it's a boolean datatype.
d = "rajkumar" # Here 'd' is a variable and "rajkumar" is a value of the variable 'd' and it's a string datatype.
print(c)  # Here we're printing the value of the variable 'c' .
print(d)  # Here we're printing the value of the variable 'd' .

```

```python
# We can also use separator to separate the variable in print function .

name = "Rajkumar"  # Here 'name' is a variable and "Rajkumar" is a value of the variable 'name' and it's a string datatype.
print("Welcome", name) # Here we're printing the value of the variable 'name' with the string "Welcome".
```

```python
# We can also use end parameter to print the variable in one line and  another variable in another line .
# Operator inside python is used to perform the operation on the variable like addition, subtraction, multiplication, division etc.

a = 1  # Here 'a' is a variable and 1 is a value of the variable 'a' and it's an integer datatype.
b = 2  # Here 'b' is a variable and 2 is a value of the variable 'b' and it's an integer datatype.
c = 3  # Here 'c' is a variable and 3 is a value of the variable 'c' and it's an integer datatype.

print(a+b+c)  # Here we're adding the value of the variable 'a', 'b' and 'c' and printing the result.
```

Two important concept inside the variable , they are : [ Mainly asked in interview]

1. Dynamic Typing 
2. Static Typing

#### 1) Dynamic Typing : 


```python
# 1) Dynamic Typing variable declaration 

# In dynamic typing, we don't have to mention the datatype of the variable which we're using in python.
    # In python, we don't have to mention the datatype of the variable, python automatically detect the datatype of the variable
    # python is a dynamically typed language and it's a high level language. 

a = 7   # Here 'a' is a variable and 7 is a value of the variable 'a' and it's an integer datatype.

print (a)  # Here we're printing the value of the variable 'a' .
```

#### 2) Static Typing : 

```python
# 2) Static Typing variable declaration 

# In static typing, we have to mention the datatype of the variable which we're using in other programming languages like C, C++, Java etc.
    # In C, C++, Java etc, we have to mention the datatype of the variable, it's a statically typed language.
    # In statically typed language, we have to mention the datatype of the variable. 
    # It will not automatically detect the datatype of the variable.

int a = 7  # Syntax for C language, C++ language, Java language etc. 

print (a)  # Here we're printing the value of the variable 'a' .
```

Two another important concept inside the variable , they are : [ Mainly asked in interview]

1. Dynamic Binding 
2. Static Binding

#### 1) Dynamic Binding :


```python
# Dynamic Binding 
# In dynamic binding, we can change the datatype of the variable at any time because python is a dynamically typed language.
    # In python, interpreter run the code line by line and it will not check the whole code at once. 

a = 5 # Here 'a' is a variable and 5 is a value of the variable 'a' and it's an integer datatype.
print(a)  # Here we're printing the value of the variable 'a' .
a = "rajkumar" # Here 'a' is a variable and "rajkumar" is a value of the variable 'a' and it's a string datatype.
print(a)  # Here we're printing the value of the variable 'a' .
```

#### 2) Static Binding : 


```python
# Static Binding 
# In static binding, we can't change the datatype of the variable at any time because C, C++, Java etc are statically typed language.
    # In C, C++, Java etc, compiler check the whole code at once and it will not run the code line by line.

int a = 5  # Here 'a' is a variable and 5 is a value of the variable 'a' and it's an integer datatype.
print(a)  # Here we're printing the value of the variable 'a' .

str a = "rajkumar"  # Here 'a' is a variable and "rajkumar" is a value of the variable 'a' and it's a string datatype.
print(a)  # Here we're printing the value of the variable 'a' .
```

Multiple variable declaration in one line

```python
# Multiple variable declaration in one line

a = 5  # Here 'a' is a variable and 5 is a value of the variable 'a' and it's an integer datatype.
b = 6  # Here 'b' is a variable and 6 is a value of the variable 'b' and it's an integer datatype.
c = 7  # Here 'c' is a variable and 7 is a value of the variable 'c' and it's an integer datatype.

print(a+b+c)  # Here we're adding the value of the variable 'a', 'b' and 'c' and printing the result.
```

```python
# Multiple variable declaration in one line

a,b,c = 5,6,7  # Here 'a', 'b', 'c' are the variables and 5,6,7 are the values of the variable 'a', 'b', 'c' and it's an integer datatype.
print(a,b,c)  # Here we're printing the value of the variable 'a', 'b', 'c' .
```

Multiple variable declaration in one line with same value : 

```python
# Multiple variable declaration in one line with same value

a = 5
b = 5
c = 5

print(a,b,c)  # Here we're printing the value of the variable 'a', 'b', 'c' .
```

```python

# Multiple variable declaration in one line with same value 

a=b=c=5  # Here 'a', 'b', 'c' are the variables and 5 is the value of the variable 'a', 'b', 'c' and it's an integer datatype.

print(a,b,c)  # Here we're printing the value of the variable 'a', 'b', 'c' .
```

# Keywords in python



- There are 39 keywords inside new version of python python  .

                 Python keywords are **reserved words** that have special meanings within the programming language. These keywords cannot be used as identifiers, such as variable names or function names, because they are integral to the syntax and structure of Python code.

**Overview of Python Keywords**

- **Definition**: Keywords are predefined words that serve specific functions in Python. They help define the syntax and control the flow of the program .
- **Case Sensitivity**: All keywords, except for `True`, `False`, and `None`, are written in lowercase and are case-sensitive .
- **Total Count**: As of Python 3.11, there are 35 keywords in total .

**Identifiers vs. Keywords :**   

                  **While keywords are reserved** and **cannot be used as identifiers**, identifiers are names given to variables, functions, classes, etc. Identifiers must follow certain rules:

- They cannot be named after keywords.
- They are case-sensitive.
- They can contain letters, digits, and underscores but must start with a letter or underscore .

                  In summary, understanding Python keywords is crucial for writing syntactically correct code and effectively utilizing the language's features.
