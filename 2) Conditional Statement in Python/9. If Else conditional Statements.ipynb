
# Conditional Statements in Python  

- **Conditional Statements** are used to execute a particular block of code based on the condition. 
- **Conditional Statements** are used to perform different computations or actions depending on whether a specific Boolean constraint evaluates to true or false. 
- **Conditional Statements** are used to make decisions in Python.
- **Conditional Statements** are used to perform different actions based on different conditions.

## Types of Conditional Statements in Python

i) **if statement**

ii) **if-else statement**

iii) **if-elif-else statement**

iv) **nested if statement**


## i) if statement
- **if statement** is used to execute a block of code if the condition is true.

```python
if condition:
    # block of code
```

- **Syntax** of **if statement** is as follows:
    - **if** condition:
        - block of code

- **Example** of **if statement** is as follows:

```python
x = 10
if x > 5:
    print("x is greater than 5")
```

- **Output** of the above code is as follows:

```
x is greater than 5
```

## ii) if-else statement

- **if-else statement** is used to execute a block of code if the condition is true and another block of code if the condition is false.

```python
if condition:
    # block of code
else:
    # block of code
```

- **Syntax** of **if-else statement** is as follows:
    - **if** condition:
        - block of code
    - **else**:
        - block of code

- **Example** of **if-else statement** is as follows:

```python   
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is less than or equal to 5")
``` 

- **Output** of the above code is as follows:

```
x is greater than 5
```

## iii) if-elif-else statement

- **if-elif-else statement** is used to execute a block of code if the condition is true and another block of code if the condition is false.

```python
if condition1:
    # block of code
elif condition2:
    # block of code
else:
    # block of code
```

- **Syntax** of **if-elif-else statement** is as follows:
    - **if** condition1:
        - block of code
    - **elif** condition2:
        - block of code
    - **else**:
        - block of code

- **Example** of **if-elif-else statement** is as follows:

```python
x = 10
if x > 5:
    print("x is greater than 5")
elif x < 5:
    print("x is less than 5")
else:
    print("x is equal to 5")
```

- **Output** of the above code is as follows:

```

x is greater than 5
```

## iv) nested if statement

- **nested if statement** is used to execute a block of code if the condition is true and another block of code if the condition is false.

```python

if condition1:
    # block of code
    if condition2:
        # block of code
    else:
        # block of code
else:
    # block of code
```

- **Syntax** of **nested if statement** is as follows:
    - **if** condition1:
        - block of code
        - **if** condition2:
            - block of code
        - **else**:
            - block of code
    - **else**:
        - block of code

- **Example** of **nested if statement** is as follows:

```python   
x = 10
if x > 5:
    print("x is greater than 5")
    if x == 10:
        print("x is equal to 10")
    else:
        print("x is not equal to 10")
else:
    print("x is less than or equal to 5")
```

- **Output** of the above code is as follows:

```
x is greater than 5
x is equal to 10
```
