```python
## Multi-line text 

text = '''Cyclone Dumazile was a strong tropical cyclone in 
the South-West Indian Ocean that affected Madagascar 
and Réunion in early March 2018. Dumazile originated 
from a Cyclone low-pressure Dyclone Pyclone area that formed 
near Agaléga on 27 February. It became a tropical 
disturbance on 2 March, and was named the next day 
after attaining tropical storm status. 
'''
```

```python
## If we have to find out "Cyclone","Dyclone","Pyclone" words from the multi-line text , 
    #  so how we'll be extracting this particular "Cyclone","Dyclone","Pyclone" words from the multi-line text.

pattern = r"[A-Z]+yclone"

# If we master the above pattern then we will become a champion in the regular expression. So, we only need to know how to write the patten in the regular expression.

## Explanation of the above code:
    # [A-Z] : This will match any uppercase letter from A to Z.
    # + : This will match one or more of the preceding character or group.
    # yclone : This will match the word yclone in the text which is case-insensitive.
    
    # So, the above pattern will match the words "Cyclone","Dyclone","Pyclone" from the multi-line text.

```

```python

import re           # Importing the re module to work with the regular expression in python.
```

```python
## If we need only the first word from the text then we can use the search function of the re module to search the pattern in the text.

match = re.search(pattern,text)    # This will search the pattern in the text and return the first match.
print(match)    # <re.Match object; span=(0, 7), match='Cyclone'> : This will return the match object which contains the information about the match.

# inside re module we have a search function which is used to search the pattern in the text.

## Explanation of the above code: 
    # re.search(pattern,text) : This will search the pattern in the text and return the first match.
    # print(match) : This will print the match object which contains the information about the match.
    
    # So, the above code will search the pattern in the text and return the first match.
    
    # If the pattern is found in the text then it will return the match object otherwise it will return None.

## Output : <re.Match object; span=(0, 7), match='Cyclone'>
    # This will return the match object which contains the information about the match like span, match.
    # span=(0, 7) : This will return the starting and ending index of the match in the text.
    # match='Cyclone' : This will return the matched word from the text which is "Cyclone".
```

```python
## If we want multiple words then, we need to use another function called finditer() of the re module to find all the matches in the text.

match_item = re.finditer(pattern,text)    # This will find all the matches in the text and return the iterator of the match objects which contains the information about the match.
                                            # re module has a finditer() function which is used to find all the matches in the text.
                                            # finditer() function will return the iterator of the match objects which contains the information about the match.
                                            # match_item is the iterator of the match objects which contains the information about the match.

for match in match_item:      # This will iterate through the match objects which contains the information about the match in the text.
    # print(match.span())     # This will print the starting and ending index of the match in the text.
    index = match.span()      # This will return the starting and ending index of the match in the text.
                                # match.span() will return the starting and ending index of the match in the text.
                                # index is the tuple which contains the starting and ending index of the match in the text.
    print(text[index[0]:index[1]])  # This will print the matched word from the text using the starting and ending index of the match.
                                        # text is the text from which we are extracting the matched word.
                                        # index[0] is the starting index of the match in the text and it is inclusive.
                                        # index[1] is the ending index of the match in the text and it is exclusive.
                                        # text[index[0]:index[1]] will return the matched word from the text using the starting and ending index of the match.

## Explanation of the above code:
    # re.finditer(pattern,text) : This will find all the matches in the text and return the iterator of the match objects which contains the information about the match.
    # for match in match_item : This will iterate through the match objects which contains the information about the match in the text.
    # match.span() : This will return the starting and ending index of the match in the text.
    # index = match.span() : This will return the starting and ending index of the match in the text.
    # print(text[index[0]:index[1]]) : This will print the matched word from the text using the starting and ending index of the match.
    
    # So, the above code will find all the matches in the text and print the matched word from the text using the starting and ending index of the match.
    
    # If the pattern is found in the text then it will return the match object otherwise it will return None.

```
